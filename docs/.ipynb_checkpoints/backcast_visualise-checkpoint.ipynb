```python
import plotly.graph_objects as go

print ("Deploying WasteNot would have had the following impact between",json_data['simulation_first_datetime'],"to",json_data['simulation_last_datetime'])

fig = go.Figure()

fig.add_trace(go.Indicator(
    mode = "number+delta",
    value = json_data['simulation_profit'],
    number = {'prefix': "$"},
    title = {"text": "Profit"},    
#     showline=True,    
#     title = {"text": "Accounts<br><span style='font-size:0.8em;color:gray'>Subtitle</span><br><span style='font-size:0.8em;color:gray'>Subsubtitle</span>"},
    delta = {'reference': json_data['historical_profit'], 'relative': False, 'position' : "right"},
    domain = {'x': [0,0.3], 'y': [0.6, 1]}
))

fig.add_trace(go.Indicator(
    mode = "number+delta",
    number = {'prefix': "$"},
    value = json_data['simulation_revenue'],
    domain = {'x': [0.35, 0.65], 'y': [0.6, 1]},

    title = {"text": "Revenue"},
    delta = {'reference': json_data['historical_revenue'], 'relative': True, 'position' : "right"}))

fig.add_trace(go.Indicator(
    mode = "number+delta",
    number = {'prefix': "$"},
    value = json_data['simulation_cost'],
    delta = {'reference': json_data['historical_cost'], 'relative': True, 'position' : "right",
             'increasing.color':"red",
             'decreasing.color':"green",},
    title = {"text": "Cost"},
    domain = {'x': [0.7, 1.], 'y': [0.6, 1.0]}
))

fig.add_trace(go.Indicator(
    mode = "number+delta",
    value = json_data['simulation_wasted_units_total'],
    title = {"text": "Wasted Units"},
#     title = {"text": "Accounts<br><span style='font-size:0.8em;color:gray'>Subtitle</span><br><span style='font-size:0.8em;color:gray'>Subsubtitle</span>"},
    delta = {'reference': json_data['historical_wasted_units_total'], 'relative': True, 'position' : "right",
             'increasing.color':"red",
             'decreasing.color':"green",},
    domain = {'x': [0,0.4], 'y': [0, 0.4]}
))

fig.add_trace(go.Indicator(
    mode = "number+delta",
    value = json_data['simulation_satisfied_units_total'],
    title = {"text": "Satisfied Units"},
#     title = {"text": "Accounts<br><span style='font-size:0.8em;color:gray'>Subtitle</span><br><span style='font-size:0.8em;color:gray'>Subsubtitle</span>"},
    delta = {'reference': json_data['historical_demand_units_total'], 'relative': True, 'position' : "right",
             'increasing.color':"red",
             'decreasing.color':"green",},
    domain = {'x': [0.6,1.0], 'y': [0, 0.4]}
))



# fig.update_layout(paper_bgcolor = "lightgray")

fig.show()
```
